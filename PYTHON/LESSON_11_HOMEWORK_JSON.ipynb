Import data to a dictionary

Calculate, looping through every student data:

a) Average Age of all students

b) average Grade of all students

c) (optional) calculate average Grade by Majo



```python
import json

with open('/content/students.json') as file:
    data = json.load(file)

total_age = 0
total_grade = 0

for student in data['Students']:
  total_age += int(student["Age"])
  total_grade += int((student["Grade"]))

avarage_age = total_age/(len(data['Students']))
avarage_grade = total_grade/(len(data['Students']))


print(f"Avarage age of all students: {avarage_age:.2f}")
print(f"Avarage grade 2 of all students: {avarage_grade:.2f}")
```
