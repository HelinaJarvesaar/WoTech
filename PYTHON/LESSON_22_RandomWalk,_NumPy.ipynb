Simulate a 1D simple random walk where at each step can be moving only up or down. How many steps ‚è´  can you actually get?

```python
import random

steps = 1000
steps_up = 0
location = 0
locations = [location]

for i in range (steps):
  step = random.choice([-1,1])
  location += step
  locations.append(location)

  if step == 1:
    steps_up += 1

print(f"Location after {steps} steps: {location}")
print(f"Number of steps-up: {steps_up}")
```

```python
import pylab

pylab.figure(figsize=(10, 6))
pylab.plot(locations, label="1D Random Walk", color='navy')
pylab.title(f"1D Random Walk - {steps} Steps")
pylab.xlabel("Steps")
pylab.ylabel("Location")
pylab.grid(True)
pylab.legend()

pylab.show()
```
