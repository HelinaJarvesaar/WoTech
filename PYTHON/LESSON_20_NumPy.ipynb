```python
import numpy as np
```

```python
coin =["Heads", "Tails"] # this is 1D array or list
```

```python
coin
```

```python
number_of_throws = 1000 # size of random list
```

```python
tosses = np.random.choice(coin, number_of_throws)
```

```python
print(f'The results of coin toss simuation for{number_of_throws} throws are {tosses}')
```

```python
# numpy.sum(a, axis=None, dtype=None, out=None, keepdims=<no value>, initial=<no value>, where=<no value>)

# Number of Heads
nr_heads = np.sum(tosses == 'Heads')
nr_heads
```

```python
# Number of Tails
nr_tails = np.sum(tosses == 'Tails')
nr_tails
```

```python
# Number of probability
prob_heads = nr_heads /number_of_throws
prob_heads
```

```python
# Number of probability
prob_tails = nr_tails / number_of_throws
prob_tails
```

```python
print(f'Probability of heads: {prob_heads}')
print(f'Probability of tails: {prob_tails}')
```

## Random - gives float values

```python
coordinates = np.random.rand(2)
x = coordinates[0] # first value of array
y = coordinates[1] # second value of array

import matplotlib.pyplot as plt

plt.scatter(x, y, s=100, color = 'green')
plt.title('Randomly selected coordinates')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

## Look uniform and normal distribution

random.rand - gives uniform distr.

random.randn - gives normal distr. (selected values are about near MEAN!)

```python
coordinates2 = np.random.rand(100000, 2)

x = coordinates2[:, 0] # we select all rows and all values from first column
y = coordinates2[:, 1] # we select all rows and all values from second column

import matplotlib.pyplot as plt

plt.scatter(x, y, color = 'blue')
plt.title('Randomly selected coordinates - uniform distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

```python
coordinates3 = np.random.randn(100000, 2)

x = coordinates3[:, 0] # we select all rows and all values from first column
y = coordinates3[:, 1] # we select all rows and all values from second column

import matplotlib.pyplot as plt

plt.scatter(x, y, s=1, c=x, cmap='rainbow')
plt.title('100000 Randomly Selected Coordinates')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.colorbar(label='X value')

plt.show()
```
