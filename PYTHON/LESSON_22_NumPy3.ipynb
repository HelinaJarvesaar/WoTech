### BROADCASTING

Allows oprations betwwen arrays of different shapes.

Broadcasting allows NumPy to automatically expand the smaller array to match the larger array.

```python
# matrix with studets scores

scores = np.array([
    [60, 89, 78, 55],
    [55, 76, 78, 32],
    [99, 89, 67, 77]
])
```

```python
scores.shape # 2D array, 3 rowsm 4 columns
```

```python
bonuses = np.array([5, 3, 10, 2])
```

```python
bonuses.shape # 1D array
```

```python
# broadcasting
adjusted_scores = scores + bonuses
adjusted_scores
```

```python
bonuses2 = np.array([5, 3, 10, 0]) # ei toimi kui on vähem väärtusi kui arrays, sp lisame 0 väärtuse
bonuses2.shape
```

```python
adjusted_scores2 = scores + bonuses2
adjusted_scores2
```

# RANDOM WALK

```python
n = 100000 # defining number of steps

x = np.zeros(n)
y = np.zeros(n)
```

```python
x
```

```python
y
```

```python
import random
value = random.randint(1,4) # 4 because we can move front, back, right, left
value
```

```python
for i in range(1, n):
  value = random.randint(1,4)

  if value == 1:
    x[i] = x[i - 1] + 1 # dot moves x coordinate +1
    y[i] = y[i - 1]     # y coordinate stayes the same
                        # [i-1] because dot start from previous position

  elif value == 2:
    x[i] = x[i - 1] - 1 # dot moves x coordinate -1
    y[i] = y[i - 1]         # y coordinate stayes the same

  elif value == 3:
    x[i] = x[i - 1]     # x coordinate stayes the same
    y[i] = y[i - 1] + 1 # dot moves y coordinate +1

  else:
    x[i] = x[i - 1]     # x coordinate stayes the same
    y[i] = y[i - 1] - 1 # dot moves y coordinate -1
```

```python
import pylab
pylab.plot(x,y)
plt.title(f'Random walk of n={n} steps')
```

```python
plt.figure(figsize=(10, 6))  # Increase figure size
plt.plot(x, y, color='blue', lw=2, alpha=0.7)  # Line with transparency
plt.scatter(x[0], y[0], c='green', s=100, label='Start')  # Mark start point
plt.scatter(x[-1], y[-1], c='red', s=100, label='End')  # Mark end point

plt.title('2D Random Walk', fontsize=16)
plt.xlabel('X-Coordinate')
plt.ylabel('Y-Coordinate')
plt.grid(True)
plt.legend()

plt.show()
```
