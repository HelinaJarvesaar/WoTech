##COMBINE 2 files

The file births.txt contains the number of births in Estonia for a
certain time period by months, while the file deaths.txt contains the
number of deaths.

- Write a program that finds and outputs the months when the natural
population growth was positive and its size.

- Update your program so that it also finds and outputs the natural
population growth for the entire period covered in the files.

```python
#births
with open("/content/2024_06_14_births.txt", "r") as file:
  births = file.readlines()   # this is a list
```

```python
#deaths
with open("//content/2024_06_14_deaths.txt", "r") as file:
  deaths = file.readlines()   # this is a list
```

```python
# len(births) == len(deaths)
for i in range(len(births)):
  birthline = births[i]
  deathline = deaths[i]

  birthnumber = int(birthline[4:])
  deathnumber = int(deathline[4:])

  growth = birthnumber - deathnumber
  month = birthline[:3]
  if growth > 0:
    print(month, growth)
```

### Changing so that column has correct indexis"

```python
import pandas as pd

births = pd.read_csv("/content/2024_06_14_births.txt", header = None, delimiter = " ") # delimiter= space

birth_column_names = ["Month", "nr.Births"]

births.columns = birth_column_names
births # this is births dataframe

deaths = pd.read_csv("//content/2024_06_14_deaths.txt", header = None, delimiter = " ") # delimiter= space

death_column_names = ["Month", "nr.Deaths"]

deaths.columns = death_column_names
deaths # this is death dataframe
```

## How to merge files to one

```python
# we create new column
births ["nr.Deaths"] = deaths["nr.Deaths"]
births ["Difference _between_births_and_deaths"] = births ["nr.Births"] - deaths ["nr.Deaths"]
births
```

## Exporting

```python
file_name = "/content/output.csv"

births.to_csv(file_name, index=False)
```
