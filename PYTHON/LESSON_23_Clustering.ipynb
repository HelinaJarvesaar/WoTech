## Unsupervised training example (clustering)

https://scikit-learn.org/dev/api/sklearn.datasets.html

```python
from sklearn.datasets import make_blobs # creating sythetic dataset
```

```python
X, y = make_blobs(n_samples=1000, centers=5, cluster_std=0.80)
```

```python
X[:, 0] # X coordinate (all values from 1st)
X[:, 1] # y coordinate (all values form 2nd)
y # the cluster of the point
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0],X[:, 1], c='grey', edgecolors ='k' )
```

## Training the model

(ps! it is also possible to visualize decision boundries:
https://stackoverflow.com/questions/22294241/plotting-a-decision-boundary-separating-2-classes-using-matplotlibs-pyplot)

```python
from sklearn.cluster import KMeans # import model
```

```python
kmeans = KMeans(n_clusters=5) # initialize the model (choose number of groups)
```

```python
kmeans.fit(X) # fit the model = train the model
```

```python
kmeans.predict(X) # do the prediction (so many numbers we defined in initzialise n_clusters)
```

```python
cluster_labels = kmeans.predict(X)
plt.scatter(X[:, 0],X[:, 1], c=cluster_labels, edgecolors ='k' )

kmeans.cluster_centers_[:, 0]
kmeans.cluster_centers_[:, 1]
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='*', s=300, color="red")
```
